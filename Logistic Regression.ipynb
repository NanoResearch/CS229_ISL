#### Basics - 

Logistic regression is a classification method, meaning that it will compute a class label.

It is also discriminatative. It will try to learn the output given the features, $P(y \mid x)$.

Using regression model, we predict a negative probability of default for balances close to zero.

For categorial responses, we want a funtion, $h_{\theta}(x)$, that outputs values between 0 and 1.

We define a set of probabilistic assumptions for our model, as shown below:

$$ P(y=1 \mid x;\theta) = h_{\theta}(x) = g(\theta^T x) = \frac{1}{1+e^{-\theta^T x}}$$

$$ P(y=0 \mid x;\theta) = 1 - h_{\theta}(x) $$ 

This is written compactly as:

$$ P(y \mid x;\theta) = h_{\theta}(x)^y (1-h_{\theta}(x))^{1-y} $$ 

#### Decision boundary - 

$h_{\theta}(x)$ is a logistic funtion that outputs values between 0 and 1.

$h_{\theta}(x)=0.5$ is the decision boundary.

This is produced when $\theta^Tx = \theta_1 x_1 + \theta_2 x_2 + \theta_3 = 0$.

In turn, we can define a line of values for $\theta_{1,2}$ over which the decision is ambiguous.

$x_2 = -\frac{\theta_1}{\theta_2} x_1 + \theta_3 $

#### Learning strategy: Maximize likelihood function with gradient ascent - 

Compress the two equations above, resulting in a likelihood function:

$$ L(\theta) = P(y \mid x;\theta) = (h_{\theta}(x))^y (1 - h_{\theta}(x))^{1-y} $$ 

$$ l(\theta) = log L(\theta) = \sum\limits_{i=1}^m y^i \times log h_{\theta}(x^i) + (1-y^i) \times log(1- h_{\theta}(x^i)) $$

As with $J(\theta)$ for linear regression, $l(\theta)$ for logistic regression is convex.

In turn, we can use gradient ascent to iterativly update $\theta$: 

$ \theta_j := \theta_j + \alpha \frac{\partial}{\partial \theta_j} l(\theta) $ where the update is simultanously performed for all values of $j$.

This solves to:

$$ \theta_j := \theta_j + \alpha (y^i - (h_{\theta} (x^i)) x_j^i $$

Recall for linear regression: $h_{\theta}=\theta^Tx$.

Now, for logistic regression: $h_{\theta}=\frac{1}{1+e^{-\theta^T x}}$.

The classic way to visualize a distribution is a histogram, which is shown along with a rugplot. 

```python
q=d.default=='No'
d.default=1
d.default[q]=0
```

```python
sns.set_context(rc={"figure.figsize": (10, 6)})
sns.set_palette("hls", 1)
q=d.default==1
plt.hist(np.array(d.balance[q]),alpha=.3,bins=10)
sns.rugplot(d.balance[q])
```

In the histogram, each tick is simply binned and bin frequencies are plotted.

The kernel density plot (below) is also a transformation from tick to a height-encoded measure of density. 

Instead of binning each tick mark, it represents each tick with a gaussian basis function.

We then estimate the distribution that our samples came from from by summing these basis functions.

```python
sns.kdeplot(d.balance[q],shade=True)
sns.rugplot(d.balance[q])
```

We can exaimine this using the Wilcoxon rank-sum test.

This tests the null hypothesis that two sets of measurements are drawn from the same distribution.



```python
import scipy.stats
q=d.default==1
default=d.balance[q]
nodefault=d.balance[~q]
p_value=scipy.stats.ranksums(default,nodefault)[1] 
print p_value
```

The resulting P-value suggests that the Null hypothesis is false in this case; the partitions are drawn from different distributions.

Let's use a scikit-learn estimator to train a logistic model, which will predict default based upon balance using this data.

```python
sns.set_context(rc={"figure.figsize": (10,5)})

# Linear model -
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression() # estimator that implements the methods, fit and predict

feat=np.array([d.balance]).transpose()
targ=np.array([d.default]).transpose()

clf.fit(feat,targ)
predicted=clf.predict(feat)
plt.scatter(d.balance,d.default,c='b',alpha=0.1)
plt.scatter(feat,predicted,c='r',alpha=0.1)

b=clf.coef_[0][0]
bo=clf.intercept_[0]

r=np.linspace(0,3000,100)
h=1/(1+1/np.exp(b*r+bo))
plt.plot(r,h,c='r',alpha=0.95)

plt.xlabel('Balance')
plt.ylabel('Default')
```

Let's check the score.

```python
clf.score(feat,targ)
```
