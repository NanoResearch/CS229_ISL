#### Basics -

Linear regression will compute a real number.

It is also discriminatative. It will try to learn the output given the features, $P(y \mid x)$.

$Y$ is s scalar whereas $X$ is a vector of $n$ features.

We assume that a parametric linear model captures the relationship between $Y$ and $X$:

$Y = h_{\theta}(X) = \sum\limits_{i=0}^n \theta_i x_i = \theta^T X $. 


#### Assumptions - 

The relationship between the predictors and response are additive and linear. 

Additive: The effect of changes in a predictor $X_j$ is independent of the values of the other predictors.

Linear: Change in the response Y due to a one-unit change in $X_j$ is constant, regardless of the value of $X_j$.

#### Learning strategy 1: Gradient Descent - 

One side is to learn parameters ($\theta$) that minimize the squared error between the model ($\theta^T X$) and measured target ($Y$) for our $m$ training examples.

In CS229, the objective is defined below (ISL defines this as residual sum of squares, or RSS):

$J(\theta) = \frac{1}{2} \sum\limits_{i=1}^m (h_{\theta} (X^i) - y^i)^2$

This cost fuction is convex quadratic, meaning it has a single global minimum.

This is shown below for a single training example using $\theta = (\beta_o,\beta_1)$ from the ISL text.

There is a point (red, above) in parameter space that minimizes the cost funtion ($RSS$, above).

The slope of the cost funtion with respect to one of the parameters, $\theta_j$, is:

$ \frac{\partial}{\partial \theta_j} J(\theta) \leq 0 $

Intuitively, this slope is negative until the global minimum is reached.

So, we can use the gradient descent algorithm:

* Start with a guess for $\theta$
* Iterativly change $\theta$ in the direction of greatest descent of $J(\theta)$

$ \theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta) $ 

We can see:

* $\alpha$ as the learning rate 
* := indicates assignment of values any element $j$ of $\theta$

As we expect, $\theta_j$ is updated until $\frac{\partial}{\partial \theta_j} J(\theta) = 0$ at the global minimum.

---

We can derive Gradient descent analytically, starting with the cost funtion: 
    
$$J(\theta) = \frac{1}{2} \sum\limits_{i=1}^m (h_{\theta} (x^i) - y^i)^2$$

Lets neglect the summation, so we will show this for one training example:
    
$$J(\theta) = \frac{1}{2} (h_{\theta} (x^i) - y^i)^2$$

We compute the partial with respect to one of our parameters, $\theta_j$:

$$ \frac{\partial J(\theta)}{\partial \theta_j} = \frac{\partial}{\partial \theta_j}\frac{1}{2} (h_{\theta} (x) - y)^2 $$

$$ = (h_{\theta} (x) - y) \frac{\partial}{\partial \theta_j} (\sum\limits_{j=1}^m \theta_j (x_j)) - y^i $$

$$ = (h_{\theta} (x) - y) x_j $$

This results in the least mean squares (Widrow-Hoff) updater rule:

$$ \theta_j := \theta_j + \alpha (y^i - (h_{\theta} (x^i)) x_j^i $$

---

In practice, we:

* Compute this for all paramters ($\theta_1$ to $\theta_n$) in parallel.
* Perform this across our training examples.

Batch:

* In batch mode, we perform each update to $\theta$ using all training examples.
* The problem is that this method can be slow for large $m$, as it must process all $m$ prior to making a step.

$$ \theta_j := \theta_j + \alpha \sum\limits_{i=1}^m (y^i - (h_{\theta} (x^i)) x_j^i $$

Stochastic:

* In stochastic mode, we first converge on $\theta$ using a single training example. 
* Then, we move to the next training example, starting with the prior converged value for $\theta$.

For $i$ in $m$:

$$ \theta_j := \theta_j + \alpha (y^i - (h_{\theta} (x^i)) x_j^i $$


#### Learning strategy 2: Normal equations - 

Rather than using an iterative algorithm, we can also try to find a closed-form solution for $\theta$ that minimizes $J(\theta)$.

After using matrix derivatives and some tricks (see CS229 notes), we end up with a simple expression:

$$ \nabla_{\theta} J(\theta) = X_T X \theta - X^T y_{tar} $$

Set this to zero in order get the value of $\theta$ at which there's no change in the cost funtion, indicating minimization:

$$ \theta = (X_T X)^{-1} X^T y_{tar} $$


#### Learning strategy 3: Maximize likelihood - 

Our model prediction deviates from actual value by an error term: 

$y^i = \theta^T x^i + \epsilon^i$.

Lets $assume$ errors are IID (independently and identically distributed) according to a Gaussian distribution with mean zero and some variance $\sigma^2$.

Recall the probability density function, which tells use the relative likelihood for this random variable to take on a given value:

$$ P(x) = \frac{1}{\sigma \sqrt{2\pi}} exp(-\frac{(x-\mu)^{2}}{2\sigma^2}) $$


```python
import scipy.stats as stats
u,s=0,1
rv = stats.norm(loc=u,scale=s)   
x = np.arange(-5,5,0.1)
plt.plot(x,rv.pdf(x),lw=2,color='b')
```

Useful reminders for the Normal distribution:

(1) Compute the expected value, $ E[x] = \int\limits^{\infty}_{-\infty} x P(x) dx = \mu $

(2) Compute the z-score for a given value, $ Z=\frac{(x-\mu)}{\sigma} $

(3) Known fractions (e.g., 68%, 95%) of the data within specified z-values from mean (e.g., 1, 2).

(4) Integrate density to convert to absolute probability of with observing value within a given interval: $ P[b<x<a] = \int\limits^{a}_{b} P(x) dx $

---

Thus, we know:

$$ P(\epsilon^i) = \frac{1}{\sigma \sqrt{2\pi}} exp\frac{-(\epsilon^i)^{2}}{2\sigma^2} $$

Substituting for $\epsilon$ using our definition ($y^i = \theta^T x^i + \epsilon^i$) above:

$$ p(y^i \mid x^i;\theta) = \frac{1}{\sqrt{2 \pi \sigma}} exp \frac{ -(y^i - \theta^{T}x^i)^2 }{2 \sigma^2} $$

This simply states the probability of observing a target, $y^i$, given the features $x^i$ and set of paramters $\theta$.

Intuitivly, the probability (density) increaces as our model gets closer to the target $y^i$, minimizing the numerator.

A higher probability indicates that the chosen parameters, $\theta$, yield a model that is close of the target. 

The likelihood of $\theta$ for all training data is then the product of probabilities each example, $i$: 

$$L(\theta) = P(y_{tar} \mid X;\theta) = \prod\limits_{i=1}^m P(y^{i} \mid x^i;\theta) $$

So, we can simply find $\theta$ that maximizes $L$ (or log($L$)):

$$  l(\theta) = log L(\theta) = log \prod\limits_{i=1}^m P(y_{i} \mid x^i;\theta) = m \times log \frac{1}{\sqrt{2 \pi \sigma}} + \sum\limits_{i=1}^m - \frac{(y^i - \theta^{T}x^i)^2}{2 \sigma^2} = m \times log \frac{1}{\sqrt{2 \pi}\sigma} + \sum\limits_{i=1}^m - \frac{(y^i - h_{\theta}(x^i))^2}{2 \sigma^2} $$

Maximizing $L(\theta)$ is the same as minimizing our initial least squares cost funtion, $J(\theta) = \frac{1}{2} \sum\limits_{i=1}^m (h_{\theta} (x^i) - y^i)^2$.

Thus, under Gaussian probabilistic assumptions, we derive least-squares regression.

#### Parameter evaluation with standard error and p-Value - 

Model parameterization with the training data is a general statistical approach. 

We use information from a sample to estimate characteristics of a population.

Consiser the sample case of computing the sample mean as an estimate of a population mean:

$\hat{\mu} = \frac{1}{n} \sum\limits_{i=1}^n y_i $

If we were to average means ($\hat{\mu}$) over many samples, the result will be close to the population mean. 

But, we can compute the accuracy of a single estimate $\hat{\mu}$ with standard error:

var($\hat{\mu}$) = SE($\hat{\mu})^2$ = $ \frac{\sigma^2}{n} $

Where $\sigma$ is the stndard deviation of each training example $y_i$ in the sampled data.

This tells us the average amount that this estimate $\hat{\mu}$ differs from the actual value.

It also indicates that the error drops with respect to the number of samples data points. 

Importaintly, the $SE$ can be computed for trained paramters from linear regression.

See the ISL text for the equations. 

---

Note that $SE$ can be used to:

* Compute confidence intervals
* Compute a p-value associted with our estimtae

Confidence intervals are the range of values that contains the true unknown value of the parameter.

For linear regression, the 95% CI of our parameter is $2 SE(\hat\beta_1)$.

---

We can also perform a hypothesis test. 

We ask whether $\hat\beta_1 = 0$, meaning no relationship between the ourput and feature $1$.

We can use the t-statistic to determine the number of $SE$ away from zero that our sampled parameter is.

$$ t = \frac{\hat\beta_1 - 0}{SE(\hat\beta_1)} $$

Using a table, we can compute the probability of observing any value equal to $t$ or larger, assuming $\hat\beta_1 = 0$. 

As expected, this p-value drops as $t$ gets larger: p of 5% and 1% correspond to t-statistics of 2 and 2.75, respectively, for $n=30$.


#### Measures of linear model fit - 

Recall the sum of the squared deviations from the mean $ TSS = \sum\limits (y^i-y_{mean})^2 $.

This is related to variance:

$$ v = \frac{TSS}{n-1} $$ 

Of course, the standard deviation is simply $ v = \sigma^2 $.

The variance is a special case of covariance, which is a measure of how two variables change together:

$$ Cov(a,b) = \frac{1}{n-1} \sum\limits_{i=1}^{n} (a_i - \mu_a) (b_i - \mu_b) $$

When the covariance is normalized by standard deviation, one obtains the correlation coefficient:

$$ r = Cor(a,b) = \frac{Cov(a,b)}{\sigma_a \sigma_b} $$

This is a measure of linear dependence between the two random variables. 

An alternative measure of linear dependence is $R^2$, is the proportion of variability in Y that can be explained using X.

$$ R^2 = 1 - \frac{RSS}{TSS} $$

RSS measures the amount of variability that is left unexplained after performing the regression = $ \sum\limits (y^i-\theta^T x^i)^2 $ 
 
In the simple (single predictor) linear regression setting, squared correlation is equivalent to $R^2$. 

This breaks if we have more than one predictor, as correlation quantifies association between a single pair of variables only.

* The linear regression model assumes that there is a straight-line relationship between the predictors and the response.
* Residual ($y_i - \hat{y_i}$)plots are a useful graphical tool for identifying non-linearity.
